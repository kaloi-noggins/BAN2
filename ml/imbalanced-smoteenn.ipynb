```python
from collections import Counter
```

```python
from sklearn.datasets import make_classification
```

```python
from imblearn.combine import SMOTEENN
```

```python
X, y = make_classification(n_samples=10000, n_classes=3, n_redundant=0,
n_clusters_per_class=1, weights=[0.85, 0.14, 0.01], flip_y=0)
```

```python
print(Counter(y))
```

```python
sample = SMOTEENN(sampling_strategy={0: 8500, 1: 8500, 2: 8500})
```

```python
X_over, y_over = sample.fit_resample(X, y)
```

```python
print(Counter(y_over))
```
