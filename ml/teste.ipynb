```python
import pandas as pd
from sklearn.cluster import KMeans
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
df = pd.read_csv('car.data')
df
```

```python
df['u'] = OrdinalEncoder().fit_transform([[i] for i in df['unacc'].to_list()])
df
```

```python
df2 = df[['0','1','22','22.1','s','l','u']]
df2
```

```python
qt_grupos = 5

# Aplicar o algoritmo de K-Means para agrupar os clientes
kmeans = KMeans(n_clusters=qt_grupos)
```

```python
kmeans.fit(df2)

# Obter os rÃ³tulos dos clusters
labels = kmeans.labels_
```

```python
df2['grupo'] = labels
df2
```

```python
df['grupo'] = labels
df
```
