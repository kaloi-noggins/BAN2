```python
import pandas as pd
from sklearn.linear_model import LinearRegression
```

## Previsão do preço de uma casa com base em seus recursos, como tamanho e número de quartos

```python
# Criar um DataFrame com dados de casas
data = {'area': [1200, 1500, 1000, 1800],
        'quartos': [3, 4, 2, 4],
        'preco': [300000, 450000, 250000, 550000]}
df = pd.DataFrame(data)
df
```

```python
# Separar os recursos (X) e o alvo (y)
X = df[['area', 'quartos']]
y = df['preco']
```

```python
# Treinar o modelo de regressão linear
model = LinearRegression()
model.fit(X, y)

# Prever o preço de uma nova casa
new_house = pd.DataFrame({'area': [1600], 'quartos': [3]})
predicted_price = model.predict(new_house)
print(f"Previsão de valor para a casa com area {new_house.area.item()} e com {new_house.quartos.item()} quartos:")
print("R$ {:,.2f}".format(predicted_price[0]).replace(",", "v").replace(".", ",").replace("v", "."))
```
