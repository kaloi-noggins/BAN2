```python
import pandas as pd
from sklearn.svm import SVC
```

### Classificação de documentos em categorias, como esportes, política ou entretenimento, com base em seu conteúdo textual.

```python
# Criar um DataFrame com dados de classificação de documentos
data = {'documento': ['Esporte', 'Política', 'Entretenimento'],
        'palavra1': [1, 0, 1],
        'palavra2': [0, 1, 1]}
df = pd.DataFrame(data)
df
```

```python
# Separar os recursos (X) e a categoria (y)
X = df[['palavra1', 'palavra2']]
y = df['documento']
```

```python
model = SVC()
```

C-Support Vector Classification.

```python
# Treinar o modelo de SVM
model.fit(X, y)

# Classificar um novo documento em uma categoria
new_document = pd.DataFrame({'palavra1': [1], 'palavra2': [0]})
predicted_category = model.predict(new_document)
print(predicted_category[0])
```
