````python
import pandas as pd
from sklearn.neural_network import MLPClassifier
````

### Reconhecimento de Tipos manuscritos em imagens, em que uma rede neural é treinada para identificar corretamente os números escritos.

````python
# Criar um DataFrame com dados de reconhecimento de tipos
data = {'imagem': ['imagem1', 'imagem2', 'imagem3'],
        'pixel1': [0.2, 0.4, 0.1],
        'pixel2': [0.3, 0.1, 0.5],
        'tipo': [3, 5, 2]}
df = pd.DataFrame(data)
df
````

````python
# Separar os recursos (X) e o alvo (y)
X = df[['pixel1', 'pixel2']]
y = df['tipo']
````

````python
model = MLPClassifier()
````

Multi-layer Perceptron classifier.

This model optimizes the log-loss function using LBFGS or stochastic gradient descent.

```hidden_layer_sizes=(100,), activation='relu', *, solver='adam', alpha=0.0001, batch_size='auto', learning_rate='constant', learning_rate_init=0.001, power_t=0.5, max_iter=200, shuffle=True, random_state=None, tol=0.0001, verbose=False, warm_start=False, momentum=0.9, nesterovs_momentum=True, early_stopping=False, validation_fraction=0.1, beta_1=0.9, beta_2=0.999, epsilon=1e-08, n_iter_no_change=10, max_fun=15000```

Doc: https://scikit-learn.org/stable/modules/generated/sklearn.neural_network.MLPClassifier.html

````python
# Treinar o modelo de RNA
model.fit(X, y)

# Reconhecer o tipo em uma nova imagem
new_ = pd.DataFrame({'pixel1': [0.15], 'pixel2': [0.25]})
predicted = model.predict(new_)
print(f'Tipo previsto: {predicted[0]}\n\n')
````
