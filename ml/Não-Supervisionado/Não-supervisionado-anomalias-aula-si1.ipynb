````python
import pandas as pd
from sklearn.ensemble import IsolationForest
````

Monitoramento de Rede: Em segurança de rede, o Isolation Forest pode ser usado para detectar atividades de rede anômalas, identificando comportamentos incomuns ou ataques cibernéticos.

ou ainda

Manutenção Preditiva: Em setores industriais, o Isolation Forest pode ser aplicado para detectar anomalias em sensores e equipamentos, ajudando na manutenção preditiva ao identificar falhas potenciais antes que ocorram.

````python
# Criar um DataFrame com dados de transações
data = {'valor': [100, 200, 300, 400, 500, 1000]}
df = pd.DataFrame(data)
df
````

## Isolation Forest Algorithm.

Return the anomaly score of each sample using the IsolationForest algorithm

The IsolationForest ‘isolates’ observations by randomly selecting a feature and then randomly selecting a split value between the maximum and minimum values of the selected feature.

```class sklearn.ensemble.IsolationForest(*, n_estimators=100, max_samples='auto', contamination='auto', max_features=1.0, bootstrap=False, n_jobs=None, random_state=None, verbose=0, warm_start=False)```

DOC: https://scikit-learn.org/stable/modules/generated/sklearn.ensemble.IsolationForest.html

````python
# Aplicar a técnica de Isolation Forest para detecção de anomalias
isolation_forest = IsolationForest(contamination=0.1)
````

contamination=‘auto’ or float, default=’auto’

The amount of contamination of the data set, i.e. the proportion of outliers in the data set. Used when fitting to define the threshold on the scores of the samples.

If float, the contamination should be in the range (0, 0.5).

````python
isolation_forest.fit(df)

# Obter as previsões de anomalia (-1 para anomalias, 1 para dados normais)
predictions = isolation_forest.predict(df)
predictions
````

````python
df['é-Normal?'] = ['Anomalia' if i < 0 else 'Normal' for i in predictions]
df
````
