```python
import pandas as pd
from mlxtend.frequent_patterns import apriori, association_rules
```

Pode ser utilizado para Recomendação de Produtos: Os algoritmos de regras de associação podem ser aplicados em sistemas de recomendação para identificar padrões de compra e sugerir itens relacionados a um cliente com base em seu histórico de compras.
ou "simplesmente para o planejamento de Layout" de Loja: Em negócios de varejo, as regras de associação podem ser usadas para identificar quais produtos têm alta probabilidade de serem comprados juntos. Isso pode ajudar a otimizar o layout da loja, colocando esses produtos próximos uns dos outros para incentivar vendas adicionais.


```python
# Criar um DataFrame com dados de compras
data = {'produto1': [1, 1, 0, 1, 0],
        'produto2': [1, 0, 1, 1, 1],
        'produto3': [0, 1, 1, 1, 0]}
df = pd.DataFrame(data)
df
```

Get frequent itemsets from a one-hot DataFrame

apriori(df, min_support=0.5, use_colnames=False, max_len=None, verbose=0, low_memory=False)

DOC: https://rasbt.github.io/mlxtend/api_subpackages/mlxtend.frequent_patterns/

```python
# Gerar os itens frequentes usando o algoritmo Apriori
frequent_items = apriori(df, min_support=0.2, use_colnames=True)
```

Generates a DataFrame of association rules including the metrics 'score', 'confidence', and 'lift'

association_rules(df, metric='confidence', min_threshold=0.8, support_only=False)

DOC: https://rasbt.github.io/mlxtend/api_subpackages/mlxtend.frequent_patterns/#association_rules

```python
# Gerar as regras de associação
rules = association_rules(frequent_items, metric="confidence", min_threshold=0.5)

# Exibir as regras de associação
rules
```

**antecedents**: Mostra os antecedentes de cada regra de associação, ou seja, o conjunto de itens que são associados a um determinado resultado.

**consequents**: Indica os consequentes de cada regra de associação, ou seja, o item ou conjunto de itens que estão associados aos antecedentes.

**support**: É a medida de frequência relativa com que a regra de associação ocorre no conjunto de dados.

**confidence**: Refere-se à probabilidade condicional de o consequente ocorrer, dado que o antecedente já ocorreu.

**lift**: Representa a razão entre a confiança da regra e a probabilidade de o consequente ocorrer independentemente do antecedente. O lift maior que 1 indica uma associação positiva, indicando uma maior probabilidade de o consequente ocorrer quando o antecedente está presente.

**leverage**: Mede o desvio da ocorrência conjunta dos antecedentes e consequentes em relação à ocorrência esperada se fossem independentes. Um valor positivo indica uma ocorrência conjunta maior do que o esperado.

**conviction**: Mede a dependência assimétrica entre o antecedente e o consequente, levando em consideração a probabilidade de o consequente ocorrer sem o antecedente. Valores de convicção maiores que 1 indicam uma dependência positiva.
